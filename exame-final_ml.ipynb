```python
zero_class = df[df['Heart_Attack_Incidence'] == 0]
one_class = df[df['Heart_Attack_Incidence'] == 1]

zero_class_downsampled = zero_class.sample(n=len(one_class), random_state=42)

balanced_df = pd.concat([one_class, zero_class_downsampled])

balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
cat_columns = []
scalar_columns = []
numeric_columns = []

for col in df.columns:
    if df[col].dtypes == 'int64':
        numeric_columns.append(col)
    elif df[col].dtypes == 'float64':
        scalar_columns.append(col)
    elif df[col].dtypes == 'object':
        cat_columns.append(col)
```
