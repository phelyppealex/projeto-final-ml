```python
# Checando balanceamento da variável target
df['Heart_Attack_Incidence'].value_counts()
```

```python
# Fazendo a separação entre valores positivos e negativos
zero_class = df[df['Heart_Attack_Incidence'] == 0]
one_class = df[df['Heart_Attack_Incidence'] == 1]

# Fazendo downsampling
zero_class_downsampled = zero_class.sample(n=len(one_class), random_state=42)

# Concatenando novamente o dataframe
balanced_df = pd.concat([one_class, zero_class_downsampled])

# Reembaralhando, para não haver viés
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# Checando balanceamento dos dados a cada ano (antes do downsampling)
df['Year'].value_counts()
```

```python
# Checando balanceamento dos dados a cada ano (após o downsampling)
balanced_df['Year'].value_counts()
```

```python
# Checando dados nulos
balanced_df.isnull().sum()
```

```python
# Separando variáveis escalares, inteiras e categóricas
cat_columns = []
scalar_columns = []
numeric_columns = []

for col in df.columns:
    if df[col].dtypes == 'int64':
        numeric_columns.append(col)
    elif df[col].dtypes == 'float64':
        scalar_columns.append(col)
    elif df[col].dtypes == 'object':
        cat_columns.append(col)
```

```python
# !pip install scikit-learn

from sklearn.preprocessing import OrdinalEncoder, StandardScaler

enc = OrdinalEncoder()

# Codificando colunas categóricas
balanced_df[cat_columns + numeric_columns] = enc.fit_transform(balanced_df[cat_columns + numeric_columns])
```

```python
# Verificando colunas codificadas
balanced_df
```

```python
scaler = StandardScaler()

# Padronizando variáveis escalares
balanced_df[scalar_columns] = scaler.fit_transform(balanced_df[scalar_columns])
```
